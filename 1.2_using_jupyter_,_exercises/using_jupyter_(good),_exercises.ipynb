```python
%%timeit

squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

The code was ran 1,000,000 times (referred to as 1,000,000 loops), 7 times.

$\pm$ 52.7 ns refers to the standard devitation within the 7 times to run the 1 loop.

where ns is nanoseconds
