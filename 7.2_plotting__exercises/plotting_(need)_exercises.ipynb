```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='brown', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y value with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

```python
x = np.arange(1, 51)

function1=(np.exp(-x/10))
function2=(np.sin(x))
functionf= function1 * function2

plt.plot(x, functionf, color='blue')
plt.grid(alpha=.25)
plt.xlabel('x', fontsize='15', color='darkorange')
plt.ylabel('y', rotation=0, fontsize='15',color='darkorange')
plt.title(r'$y = e^{-x/10}sin(x)$', fontsize=20)
plt.ylim(-1, 1)
```
