```python
import os
import glob
import shutil
import math
import numpy as np
```

```python
month_conversion={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}


example = "oct-08_singapore.txt"

# below code splits up the date into 3 parts, month_day and city first, then further splits it
# into day and month by differentiating them by the seperating symbol (ie - and _) the 1
# placed next to the underscore is 
month_day, city_ext = example.split("_")
month, day = month_day.split("-")

# defining a new variable using month_num, which takes the dictionary above, and compares the
# month value which has been separated above to the dictionary and replaces it accordingly. 
month_num = month_conversion.get(month)

# the below simply pieces the separated elements back together
new_name = f"{month_num}-{day}-{city_ext}"

#overall, the code is quite simple in the fact that it just takes the file name, breaks it down
#into 3 parts, focuses on the month, replaces it, and then pieces it back together with the    
#new month format
print("Original:", example)
print("Converted:", new_name)

```

```python
def rename_my_file(old_file_name):
    
    folder = r"C:\Users\keega\Documents\GitHub\learning-portfolio-KeeganGoh\6.2_files,_folders_&_os__exercises\os-collaboration-exercise-data"
    
    #note to self that os.path.join just makes a full path which connects the folder to the path name
    old_path = os.path.join(folder, old_file_name)
    
    if old_file_name[:2].isdigit(): #this checks if it has already been converted as they have
        print(f"Skipping already converted file: {old_file_name}")#the mm format already
        return
    
    #this checks if file exists and skips it if it does
    if not os.path.exists(old_path):
        print(f"File '{old_file_name}' not found in folder '{folder}'")
        return
    
   

    else:
        #does same thing and splits and converts the month format
        month_day, city_ext = old_file_name.split("_")
        month, day = month_day.split("-")
        month_num = month_conversion.get(month)

   
        new_name = f"{month_num}-{day}-{city_ext}" #different from before this not only combines the names back
        new_path = os.path.join(folder, new_name) #but here it also recreates a new path to the folder and new name

        # Rename the file on disk
        os.rename(old_path, new_path) #does what it says it does lol
        return new_name
```

```python
file_list = np.array(['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']) 

for file_name in file_list:
    converted_name = rename_my_file(file_name)  # capture return value
    if converted_name:  # only print if file existed and was renamed
        print(f"{file_name} --> {converted_name}")
```
